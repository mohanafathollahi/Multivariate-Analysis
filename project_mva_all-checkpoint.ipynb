### Detection of outliers

**Multivariate Outlier Detection**

The classic Mahalanobis Distance is a method that measures the distance between a point and a distribution to which that point belongs. This techniquee acts as a statistical measure for the classification of a point as an outlier based on a chi-square distribution.

For many of its multivariate functions, this package uses the Robust Mahalanobis distance based the Minimum Co-variance Determination. While classic calculation of the distances is widely accepted, the robust calculation is preferred.

Outlier detection techniques will normalize all of the data, so the mismatch in scaling is of no consequence. 

```r
library(depth.plot)
library(mvoutlier)
x <- na.omit(dat.clean[,continuous])
distances <- dd.plot(x, quan=1/2, alpha=0.05)
outliers <- aq.plot(x, delta=qchisq(0.95, df = ncol(x)),quan = 1/2, alpha = 0.05)
```

The graphics above provide an intuitive visualization of the outliers contained in this dataset. ### Imputing missing data 

Try two approaches MICE and random forest separately in each of the partition. We will first do training.

**MICE (Multivariate Imputation via Chained Equations)**

MICE (Multivariate Imputation via Chained Equations), as opposed to single imputations (such as mean), accounts for the statistical uncertainty in the imputations. This methods assumes that the missing data are Missing at Random (MAR), meaning that the probability that a value is missing depends only on observed value and can be predicted using them. It imputes data on a variable by variable basis by specifying an imputation model per variable. This makes MICE a very ﬂexible method that can handle variables of varying types. 

> Single imputation procedures, such as mean imputation,are an improvement but do not account for the uncertaintyin the imputations; once the imputation is completed,analyses proceed as if the imputed values were the known,true values rather than imputed. This will lead to overlyprecise results and the potential for incorrect conclusions

Precisely, the methods used by this package are:
* PMM (Predictive Mean Matching)  – For numeric variables
* logreg(Logistic Regression) – For Binary Variables( with 2 levels)
* polyreg(Bayesian polytomous regression) – For Factor Variables (>= 2 levels)

```r
sapply(train, function(x) sum(is.na(x))) #original train

# Replace cholestero 0 with NA and imput it as missing values
train$chol[train$chol == 0] <- NA

sapply(train, function(x) sum(is.na(x))) #original train
```

```r
k= 3

init = mice(train, maxit=0, seed=500) 
meth = init$method
predM = init$predictorMatrix
meth[c("tretbps","chol","thalach","oldpeak")]="pmm" 
meth[c("fbs","exang")]="logreg" 
meth[c("restecg","slope")]="polyreg"
meth[c("age","sex","cp","num","source")]=""

imputed = mice(train, method=meth, maxit = 20, predictorMatrix=predM, m=k)

# Look for methods used
imputed$method

#Imputed dataset
train.complete = complete(imputed, 1)  #action="long"
```

```r
# Checking Convergence of the imputations -> inspect the trace lines
plot(imputed)

#Plotting and comparing values with xyplot() for continuous
xyplot(imputed, tretbps ~ chol | .imp, pch = 20, cex = 1.4)
xyplot(imputed, thalach ~ oldpeak | .imp, pch = 20, cex = 1.4)

#Density plot
densityplot(imputed)
```
